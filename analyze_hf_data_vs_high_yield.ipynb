```python
ticker_data = pd.read_pickle('data/hf_data.pkl')

ticker_data.index = pd.to_datetime(ticker_data.index)
```

```python
ticker_data
```

```python
hf_data = ticker_data['Adj Close'].pct_change().mean(axis=1).reset_index()
hf_data.columns=['Date','mean_pct_chg']
hf_data = hf_data.iloc[1:]

hf_data = hf_data[hf_data['Date'] != '02/01/2021']
```

```python
# hf_data.plot.line('Date','mean_pct_chg')
```

```python
moodys_a = pd.read_csv('data/DBAA.csv')
moodys_a.columns = ['date','bond']

moodys_a = moodys_a[moodys_a['bond'] != '.']

moodys_a['date'] = pd.to_datetime(moodys_a['date'])
moodys_a['bond'] = moodys_a['bond'].astype(float)

moodys_a['yield_pct_chg'] = moodys_a['bond'].pct_change()
```

```python
moodys_a.plot.line('date','bond')
```

```python
stock_bond = pd.merge(hf_data, moodys_a, how='left', left_on='Date', right_on='date')

stock_bond.head(5)
```

```python
stock_bond = stock_bond[(stock_bond['mean_pct_chg'].notnull())
                        & (stock_bond['yield_pct_chg'].notnull())]
```

```python
stock_bond['rolling_corr'] = stock_bond.mean_pct_chg.rolling(30).corr(stock_bond.yield_pct_chg)
```

```python
stock_bond.plot.line('Date','rolling_corr')
```
